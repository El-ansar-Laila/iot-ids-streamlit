```python
corr_matrix = data[num_cols].corr()
corr_matrix
```

```python
plt.figure(figsize=(18, 12))

sns.heatmap(corr_matrix, cmap="coolwarm", annot=True, annot_kws={"size": 6},cbar=True, yticklabels=True, mask=np.abs(corr_matrix) < 0.9)

plt.title("Correlation Heatmap")
plt.show()
```

```python
# Keep only upper triangle to avoid duplicate pairs
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

# Convert to long-form table
corr_table = (
    upper.stack()
         .reset_index()
         .rename(columns={
             "level_0": "Feature 1",
             "level_1": "Feature 2",
             0: "Correlation"
         })
         .assign(AbsCorrelation=lambda x: x["Correlation"].abs())
         .sort_values("AbsCorrelation", ascending=False)
)

# Show strongest correlations
corr_table.head(40)
```

```python
high_corr = corr_table[corr_table['AbsCorrelation']> 0.9]
for col in num_cols:
    print("*********** "+col+" ***********")
    table = high_corr[(high_corr['Feature 1'] == col) | (high_corr['Feature 2'] == col) ]
    print(table)
    print("\n")
```

```python
corr = data[num_cols].corr().abs()
features = corr.columns
to_drop = set()
kept = set()

for feature in features:
    if feature in to_drop:
        continue                #to keep the feature that contains infos of the previously deleted features

        # Find all highly correlated features with current feature
    correlated = corr.index[(corr[feature] > 0.9) & (corr.index != feature)].tolist()

        # Mark correlated features as to_drop
    for c in correlated:
        to_drop.add(c)

    kept.add(feature)
to_drop
```

```python
reduced_df = data.drop(to_drop, axis= 1)
```

```python
reduced_df.shape
```

# Preprocessing

### Handling imbalance

```python
data["Attack_type"] = data["Attack_type"].replace(
    {"MQTT_Publish": "NORMAL", "Thing_Speak": "NORMAL", "Wipro_bulb": "NORMAL"}
)
# merge the normal traffic flow to reduce imbalance 
```

```python
reduced_df["Attack_type"] = reduced_df["Attack_type"].replace(
    {"MQTT_Publish": "NORMAL", "Thing_Speak": "NORMAL", "Wipro_bulb": "NORMAL"}
)
```

```python
data['Attack_type'].unique()
```

```python
from sklearn.utils.class_weight import compute_class_weight

```

### Encoding Categorical data (One Hot Encoder)

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
```

```python
ohe = OneHotEncoder(sparse_output=False)
le = LabelEncoder()
```

```python
reduced_df['Attack_type']= le.fit_transform(reduced_df['Attack_type'])

classes=reduced_df['Attack_type'].unique()
class_weights = compute_class_weight(
    class_weight="balanced",
    classes=classes ,
    y=reduced_df['Attack_type']
)

cw = dict(zip(classes, class_weights))
cw
```

```python
categories= ['proto','service']

ohe_trans = ohe.fit_transform(reduced_df[categories])
ohe_df= pd.DataFrame(ohe_trans, columns= ohe.get_feature_names_out(categories))
reduced_df= pd.concat([ohe_df, reduced_df], axis=1)
reduced_df
```

```python
reduced_df = reduced_df.drop(['service','proto'], axis=1)
```

```python
reduced_df.dtypes.value_counts()
```

```python
# label encodin tha data DF , it will be used as an input for the tree based models 
categories= ['proto','service','Attack_type']

for cat in categories:
    data[cat]= le.fit_transform(data[cat])
```

```python
from sklearn.model_selection import StratifiedKFold, cross_validate , cross_val_score
from sklearn.metrics import (
    accuracy_score, f1_score, recall_score,
    confusion_matrix, roc_auc_score, roc_curve, classification_report )

from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier

import xgboost as xgb
import lightgbm as lgb
from catboost import CatBoostClassifier
```

```python
X_proc_train, X_proc_test, y_enc_train, y_enc_test = train_test_split(
X_preprocessed, y_enc, test_size=0.3, stratify=y_enc, random_state=1
)

X_raw_train, X_raw_test, y_raw_train, y_raw_test = train_test_split(
    X_raw, y_raw, test_size=0.3, stratify=y_raw, random_state=1
)
```

```python

models_proc = {
    "LogReg": LogisticRegression(max_iter=500, class_weight=cw, n_jobs=-1),
    "LinearSVM": LinearSVC(class_weight=cw, max_iter=3000),
    "KNN": KNeighborsClassifier(n_neighbors=5),
    "GaussianNB": GaussianNB()
}
```

```python
trained_proc = {}

for name, model in models_proc.items():
    model.fit(X_proc_train, y_enc_train)
    preds = model.predict(X_proc_test)

    print(name)
    print("Accuracy:", accuracy_score(y_enc_test, preds))
    print("F1-macro:", f1_score(y_enc_test, preds, average="macro"))
    print("Recall:", recall_score(y_enc_test, preds, average="macro"))
    print("-"*40)

    trained_proc[name] = model
```

```python
models_raw = {
    "DecisionTree": DecisionTreeClassifier(class_weight=cw, random_state=1),
    "RandomForest": RandomForestClassifier(n_estimators=100, class_weight=cw, n_jobs=-1),
    "GradientBoosting": GradientBoostingClassifier(),
    "XGBoost": xgb.XGBClassifier(
        objective="multi:softprob",
        eval_metric="mlogloss",
        n_estimators=100,
        n_jobs=-1
    ),
    "LightGBM": lgb.LGBMClassifier(
        n_estimators=100, class_weight=cw, n_jobs=-1, force_row_wise =True, verbosity=-1, random_state=1
    ),
    "CatBoost": CatBoostClassifier(
        iterations=100,
        class_weights=list(cw.values()),
        verbose=0
    )
}
 
```

```python
trained_raw = {}

for name, model in models_raw.items():
    model.fit(X_raw_train, y_raw_train)
    preds = model.predict(X_raw_test)

    print(name)
    print("Accuracy:", accuracy_score(y_raw_test, preds))
    print("F1-macro:", f1_score(y_raw_test, preds, average="macro"))
    print("Recall:", recall_score(y_raw_test, preds, average="macro"))
    print("-"*40)

    trained_raw[name] = model
```

```python
for name, model in trained_proc.items():
    preds = model.predict(X_proc_test)
    cm = confusion_matrix(y_enc_test, preds)

    plt.figure(figsize=(6,4))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=le.classes_, yticklabels=le.classes_)
    plt.title(f"Confusion Matrix - {name}")
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.show()
```

```python
for name, model in trained_raw.items():
    preds = model.predict(X_raw_test)
    cm = confusion_matrix(y_raw_test, preds)

    plt.figure(figsize=(6,4))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", yticklabels=le.classes_ )
    plt.title(f"Confusion Matrix - {name}")
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.show()
```

```python
plt.figure(figsize=(10,8))

classes = np.unique(y_enc_test)
y_bin = label_binarize(y_enc_test, classes=classes)

for name, model in trained_proc.items():
    if hasattr(model, "predict_proba"):
        probs = model.predict_proba(X_proc_test)
    else:
        probs = model.decision_function(X_proc_test)

    auc = roc_auc_score(y_bin, probs, average="macro", multi_class="ovr")
    fpr, tpr, _ = roc_curve(y_bin.ravel(), probs.ravel())

    plt.plot(fpr, tpr, label=f"{name} (AUC={auc:.2f})")

plt.plot([0,1],[0,1],'k--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve â€“ Linear & Distance Models")
plt.legend()
plt.show()
```

## CROSS-VALIDATION

```python
cv = StratifiedKFold(n_splits=3, shuffle=True, random_state=1)

print("\nCROSS VALIDATION RESULTS\n")

for name, model in models_proc.items():
    scores = cross_val_score(model, X_preprocessed, y_enc, cv=cv, scoring="f1_macro")
    print(name, "CV F1:", scores.mean())

for name, model in models_raw.items():
    scores = cross_val_score(model, X_raw, y_raw, cv=cv, scoring="f1_macro")
    print(name, "CV F1:", scores.mean())
```

## Tuning Hyperparameters

```python
from sklearn.model_selection import RandomizedSearchCV
```

#### Hyperparameters grid

```python

rf_param_grid = {
    "n_estimators": [100, 200],
    "max_depth": [None, 10, 20, 30],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 2, 4],
    "max_features": ["sqrt", "log2"]
}
```

```python
lgb_param_grid = {
    "num_leaves": [31, 63, 127],
    "max_depth": [-1, 6, 10],
    "learning_rate": [0.05, 0.1],
    "n_estimators": [100, 200],
    "min_child_samples": [20, 50],
    "subsample": [0.8, 1.0],
    "colsample_bytree": [0.8, 1.0]
}
```

#### Tuning

```python
rf = RandomForestClassifier(
    class_weight=cw,
    n_jobs=-1,
    random_state=1
)

rf_search = RandomizedSearchCV(
    rf,
    rf_param_grid,
    n_iter=15,
    scoring="f1_macro",
    cv=3,
    n_jobs=-1,
    verbose=1,
    random_state=1
)

rf_search.fit(X_raw_train, y_raw_train)
```

```python
lgb_model = lgb.LGBMClassifier(
    class_weight=cw,
    n_jobs=-1,
    force_row_wise=True,
    verbosity=-1,
    random_state=1
)

lgb_search = RandomizedSearchCV(
    lgb_model,
    lgb_param_grid,
    n_iter=15,
    scoring="f1_macro",
    cv=3,
    n_jobs=-1,
    verbose=1,
    random_state=1
)

lgb_search.fit(X_raw_train, y_raw_train)
```

#### Best parameters/score

```python
best_rf = rf_search.best_estimator_
print(rf_search.best_params_)
print("Best score: ",rf_search.best_score_)
```

```python
best_lgb = lgb_search.best_estimator_
print(lgb_search.best_params_)
print(lgb_search.best_score_)
```
